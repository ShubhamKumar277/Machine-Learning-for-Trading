### Making Function to output VWAP per month

```python
# Important: Function that returns data of prices and volume

def return_data(df, year, month):
    
    indexes = df[df.year == year][df.month == month].index
    df_new = df.loc[indexes]
    prices = df_new['Close Price']
    volume = df_new['No. of Trades']
    return prices, volume
    
```

```python
def vwap_over_months(): 
    
    df_vwap = pd.DataFrame( columns = [ 'Year', 'Month', 'VWAP per month'])
    
    for yr in range(2017, 2020):
        for mn in range(1, 13):
            price, vol = return_data(df, yr, mn)
            if price.size >= 18 and vol.size >= 18:
                vwap = np.divide(np.sum(np.multiply(price, vol)), np.sum(vol))
                df_vwap = df_vwap.append({'Year' : yr, 'Month' : mn, 'VWAP per month': vwap}, ignore_index = True )

            else:
                df_vwap = df_vwap.append({'Year' : yr, 'Month' : mn, 'VWAP per month': "Data Not Given"},  ignore_index=True )
    return df_vwap      


```

```python
df_vwap = vwap_over_months()
```

### Profit / Loss Percentage

```python
df
```

```python
def profit_loss_percentage(n):
    
    last_day_price = df['Close Price'].tail(1)
    
    last_day_index = df['Close Price'].tail(1).index
    
    chossen_day_price = df.loc[ last_day_index- n, ['Close Price']]
    
    change = np.divide(np.subtract(np.sum(last_day_price), np.sum(chossen_day_price)), np.sum(chossen_day_price))*100
    
    print("Profit/Loss percentage is:", change)
    return change

    
```

# 1.7

Add another column 'Trend' whose values are:

'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5  
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1  
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1  
'Positive' for 'Day_Perc_Change' in between 1 and 3  
'Negative' for 'Day_Perc_Change' in between -1 and -3  
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7  
'Among top losers' for 'Day_Perc_Change' in between -3 and -7  
'Bull run' for 'Day_Perc_Change' >7  
'Bear drop' for 'Day_Perc_Change' <-7  

```python
# making dataframe for "Trend"

def trend_predicter(df):
    
    df_x = pd.DataFrame(columns = ['Trend'])
    
    for i in range(0, len(df)):
        
        x = df.loc[i, ['Day_Perc_Change']]
        
        if (x <= 0.5).bool() and (x >= -0.5).bool():
            df_x = df_x.append({'Trend':'Slight or No change'}, ignore_index=True)
        
        elif (x < 1.0).bool() and (x > 0.5).bool():
            df_x = df_x.append({'Trend':"Slight positive"}, ignore_index=True)
        
        elif (x < -0.5).bool() and (x > -1.0).bool():
            df_x = df_x.append({'Trend':"Slight negative"}, ignore_index=True)
            
        elif (x < 1.0).bool() and (x > 0.5).bool():
            df_x = df_x.append({'Trend':"Slight positive"}, ignore_index=True)
        
        elif (x < 3.0).bool() and (x >= 1.0).bool():
            df_x = df_x.append({'Trend':"Positive"}, ignore_index=True)
        
        elif (x < -1.0).bool() and (x >= -3.0).bool():
            df_x = df_x.append({'Trend':"Negative"}, ignore_index=True)
        
        elif (x <= 7.0).bool() and (x > 3.0).bool():
            df_x = df_x.append({'Trend':"Among top gainers"}, ignore_index=True)
        
        elif (x < -3.0).bool() and (x >= -7.0).bool():
            df_x = df_x.append({'Trend':"Among top losers"}, ignore_index=True)
        
        elif ( x > 7.0).bool():
            df_x = df_x.append({'Trend':"Bull run"}, ignore_index=True)
        
        elif ( x < -7.0).bool():
            df_x = df_x.append({'Trend':"Bear drop"}, ignore_index=True)
            
    return df_x
```

```python
df_trend = trend_predicter(df)
```

```python
# Concatenating df_trend with df

frames = [df, df_trend]
df = pd.concat(frames, axis = 1)
```
